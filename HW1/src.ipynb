```python
import numpy as np
import matplotlib
import math
import random
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
```

```python
def gaussian_values(size, mean=0.0, std=1.0):
#     print(size, mean, var)
    return np.random.normal(mean, std, size)

def get_rewards(qvals):
    arr = [gaussian_values(1, x, 1)[0] for x in qvals]
    return arr
```

```python
num_runs = 1000
num_steps = 10000
num_actions = 10
epsilon = 0.1
rew = [0]*num_steps
```

# Using Sample Average Method

```python
rand = random.uniform(0, 1)
rew = [0]*num_steps
opt = [0]*num_steps
for i in tqdm(range(num_runs)):
    qstar = [rand]*num_actions #generating n equal qstar values to start with
    Q = [0]*num_actions
    action_counts = [0]*num_actions # to keep a count of the number of times each action is chosen
    optimal_count = 0 # to keep a count of number of times optimal action has been chosen
    for step_num in range(num_steps):
        qstar = qstar + gaussian_values(num_actions, 0, 0.01)
        optimal = np.argmax(qstar)
        prob = random.uniform(0, 1)
        if (prob <= epsilon):
            action = random.randint(0, 9)
        else:
            action = Q.index(max(Q))
        action_counts[action]+=1
        if (action == optimal):
            optimal_count += 1 
            opt[step_num] += 1
        reward = gaussian_values(1, qstar[action], 1)[0]
        Q[action] = Q[action] + (reward - Q[action])/action_counts[action]
        rew[step_num] += reward
```

```python
rew_samp_avg = [(x/num_runs) for x in rew]
opt_samp_avg = [(x/num_runs) for x in opt]
```

# Unbiased Constant-Step-Size Trick - not a part of assignment

```python
num_runs = 1000
num_steps = 10000
rand = random.uniform(0, 1)
rew = [0]*num_steps
opt = [0]*num_steps
for i in tqdm(range(num_runs)):
    qstar = [rand]*num_actions #generating n equal qstar values to start with
    Q = [0]*num_actions
    action_counts = [0]*num_actions # to keep a count of the number of times each action is chosen
    optimal_count = 0 # to keep a count of number of times optimal action has been chosen
    alpha = 0.1
    on = 0
    for step_num in range(num_steps):
        on = on + alpha*(1 - on)
        qstar = qstar + gaussian_values(num_actions, 0, 0.01)
        optimal = np.argmax(qstar)
        prob = random.uniform(0, 1)
        if (prob <= epsilon):
            action = random.randint(0, 9)
        else:
            action = Q.index(max(Q))
        action_counts[action]+=1
        if (action == optimal):
            optimal_count += 1 
        reward = gaussian_values(1, qstar[action], 1)[0]
        Q[action] = Q[action] + (alpha/on)*(reward - Q[action])
        rew[step_num] += reward
        opt[step_num] += optimal_count/(step_num+1)
```
